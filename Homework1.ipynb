```python
import numpy as np
M = np.array([[100, 150, 200], [50, 100, 150], [0, 50, 100]])
E = np.array([[20, 30, 40], [10, 20 ,30], [5, 10, 15]])
contrast_M = M * 0.5
brightless_M = M + 25
blended_M = 0.8 * M + 0.2 * E
print('M:\n', M)

print("n1. (M*0.5):\n", contrast_M)
print("n2. (M+25):\n", brightless_M)
print("n3. (0.8*M+0.2*E):\n", blended_M)

```

```python
from os import name
import numpy as np
u = np.array([8, 2, 5])
movies = {"film A (action)": np.array([9, 1, 2]), "film B (comedy)": np.array([1, 9, 8]), "film C (drama)": np.array([7, 2, 6])}
def cosin_similarity(v1, v2):
  dot_product = np.dot(v1, v2)
  norm_v1 = np.linalg.norm(v1)
  norm_v2 = np.linalg.norm(v2)
  return dot_product
similarity = {}
for name, v_movie in movies.items():
  score = cosin_similarity(u, v_movie)
  similarity[name] = score
  print(f"the some {name}: {score: .4f}")

best_movie = max(similarity, key = similarity.get)
print("-" * 30)
print(f"{best_movie} {similarity[best_movie]:.4f}")

```
